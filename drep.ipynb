```python
asr= pd.read_csv('/home/harini/features/asr.csv',header=None)
```

```python
asr.columns=['video_name','asr']
```

```python
asr
```

```python
asr['video_id']= asr['video_name'].str.split('_').str[0]
asr['video_id']= asr['video_id'].astype(int)
```

```python
import csv

# Iterate over the scenes and match them to the ASR output
for index, row in scenes.iterrows():
    asr_op= asr.loc[asr['video_id']==row['Video id']]['asr']
    asr_op= asr_op.to_string(index=False)
    asr_lines = asr_op.split("\n")
    asr_lines = asr_lines[:-1]
    scene_number = row["Scene Number"]
    start_time = row["Start Time (seconds)"]
    end_time = row["End Time (seconds)"]

    # Find the lines in the ASR output that match the scene's start and end times
    scene_lines = []
    for line in asr_lines:
        parts = line.split(":")
 
        timestamp = float(parts[0].split("-")[0])
        if timestamp >= start_time and timestamp <= end_time:
            scene_lines.append(line)
    
    # Combine the matching lines into a single transcription for the scene
    transcription = " ".join([line.split(":")[1].strip() for line in scene_lines])
    
    #write to file
    with open('scene_transcriptions.csv', 'a') as f:
        csv_writer = csv.writer(f)
        csv_writer.writerow([row['Video id'], scene_number, transcription])
```
